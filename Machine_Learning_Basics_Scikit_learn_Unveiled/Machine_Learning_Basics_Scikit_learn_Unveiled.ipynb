## Basic statistical questions

**Statistical inference: Is my data normally distributed?**

How to assess if your data is normally distributed? We can visualise a historgram, or perform a Shapiro-Wilk test.

First, let us create two data samples:


```python
import math
import numpy as np
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Create a small normally distributed dataset
norm_data = stats.norm.rvs(size = 5000)

#tCreate a uniformally distributed dataset
non_norm_data = stats.uniform.rvs( size=5000)
```

Next, let us examine the first set usign the two methods outlined:

```python
# Plot a histogram

plt.hist(norm_data, edgecolor='black', bins=100)

#Run a Shapiro-Wilk Test for normal distribution:

print(stats.shapiro(norm_data))
```

We see that the p-value for the sample is not low enough for us to reject the null hypothesis.

```python
# Plot a histogram

plt.hist(non_norm_data, edgecolor='black', bins=100)

#Run a Shapiro-Wilk Test for normal distribution:

print(stats.shapiro(non_norm_data))
```

## Machine Learning Algorithms

### Supervised Learning

Supervised learning algorithms are trained on a dataset that includes both input and output values. The algorithm is trained to learn to map the input values to the output values. For example, a supervised learning algorithm could be trained to predict the price of a house based on its features, such as the number of bedrooms, the square footage, and the location.

Some of the most common supervised learning algorithms include:

- Linear regression: A simple algorithm that predicts a continuous value based on a set of input values (Regression).
- Logistic regression: Is a statistical model algorithm that models the probability of a categorical value event based on a set of input values (Classification).
- Decision trees: A more powerful algorithm that can be used to classify or predict both continuous and categorical values.
- Random Forest: Also known as random decision forests, is an ensemble learning method for classification and regression.
- Support vector machines (SVMs): SVMs is a versatile and robust algorithm that can be used for both classification and regression tasks.
- Naive Bayes classifier: A simple but effective algorithm for classification tasks.
- k-Nearest Neighbors: Is an algorithm used in both classification and regression tasks.
- Artificial Neural Networks: Is an interconnected group of nodes, inspired by a simplification of neurons in a brain. They are widely used in classification of images for example.

#### Linear Regression

For this section, let us explore a dataset that contains details of hardware specifications fuel consumption, and C02 emission of cars:

```python
# Import required libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model


# Read the CSV file :
data = pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/FuelConsumptionCo2.csv")

data.head()
```

Some information on the rows in this dataset:

```python
data.info()
```

```python
# Let's select some features to explore more :
data = data[["ENGINESIZE","CO2EMISSIONS"]]

# PLOTTING ENGINESIZE vs CO2EMISSIONS:
plt.scatter(data["ENGINESIZE"] , data["CO2EMISSIONS"] , color="blue")
plt.xlabel("ENGINESIZE")
plt.ylabel("CO2EMISSIONS")
plt.show()
```

In order to design and train our model, we need to split our data into a labelled training set, and a test set, both containing identical information.

```python
# Generating training and testing data from our data:
# We are using 80% data for training.
train = data[:(int((len(data)*0.8)))] # The first 80% of data [0:852]
test = data[(int((len(data)*0.8))):]  # The rest after the first 80% of data [853:1066]

# Modeling:
# Using sklearn package to model data :
regr = linear_model.LinearRegression()     # Linear fit
train_x = np.array(train[["ENGINESIZE"]])  # Train data 1st. column
train_y = np.array(train[["CO2EMISSIONS"]]) # Train data 2nd. column
regr.fit(train_x,train_y) # Fit using training data. Returns coeffs linear fit

# The coefficients:
print ("coefficients : ",regr.coef_) #Slope
print ("Intercept : ",regr.intercept_) #Intercept
```

```python
# Plotting the regression line:
plt.scatter(train["ENGINESIZE"], train["CO2EMISSIONS"], color='blue')
plt.plot(train_x, regr.coef_*train_x + regr.intercept_, '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
```

Look at the data points in the scatter plot and the regression line. Visually, this line does a good job describing the trends in this data- if engine size increases, the CO2 emissions are also likely to increase.

Next, let us try adn predict CO2 emissions:

```python

# Function for predicting future values :
def get_regression_predictions(input_features,intercept,slope):
 predicted_values = input_features*slope + intercept
 return predicted_values

# Example predicting emission for future car:
my_engine_size = 3.5
estimatd_emission = get_regression_predictions(my_engine_size,regr.intercept_[0],regr.coef_[0][0])
print ("Estimated Emission :",estimatd_emission)
```

#### Principal Component Analysis

Refer to the scikit-learn documentation here:

https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

import pandas as pd
import seaborn as sns
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/clizarraga-UAD7/Datasets/main/penguins/penguins_size.csv'
penguins = pd.read_csv(url)
```

```python
penguins.head()
```

```python
columns_of_interest = ["species", "island", "culmen_length_mm", "culmen_depth_mm",
                       "flipper_length_mm", "body_mass_g", "sex"]
penguins = penguins.loc[:,columns_of_interest]
penguins = penguins.dropna()
```

```python
# select subset to only use numeric data for PCA
penguins_data=penguins.select_dtypes(np.number)
penguins_data.head()
```

```python
pca = PCA(n_components=4)
penguins_pca= pca.fit_transform(penguins_data)
```

Create a dataframe with the principal components

```python
pc_df = pd.DataFrame(data = penguins_pca ,
        columns = ['PC1', 'PC2','PC3', 'PC4'])
pc_df.head()
```

```python
sex=penguins.sex.tolist()
species=penguins.species.tolist()
```

Append the categorical data to the principal components for each penguin

```python
pc_df['sex']=sex
pc_df['species']=species
pc_df.head()
```

Using the explained_variance_ratio method, we can see how much of the variance each principal component explains. In this example, the first component explains the majority of the variance...

```python
pca.explained_variance_ratio_
```

```python
import seaborn as sns
plt.figure(figsize=(12,10))
with sns.plotting_context("notebook",font_scale=1.25):
    sns.scatterplot(x="PC1", y="PC2",
                    data=pc_df,
                    hue="species",
                    style="sex",
                    s=100)
```

Normalize the data to have a mean of zero, standard deviation of 1.

```python
random_state = 0
pca_scaled = make_pipeline(StandardScaler(),
                           PCA(n_components=4, random_state=random_state))

penguins_pc_scaled=pca_scaled.fit_transform(penguins_data)
```

With the scaled data, component 1 explains about 70% of the variance, and 2 explains about 20% of the variance.

```python
pca_scaled.named_steps['pca'].explained_variance_ratio_*100
```

```python
pc_scaled_df = pd.DataFrame(data = penguins_pc_scaled ,
        columns = ['PC1', 'PC2','PC3', 'PC4'])
pc_scaled_df['species'] = species
pc_scaled_df['sex'] = sex
pc_scaled_df.head()
```

```python
plt.figure(figsize=(12,10))
with sns.plotting_context("talk",font_scale=1.25):
    sns.scatterplot(x="PC1", y="PC2",
                    data=pc_scaled_df,
                    hue="species",
                    style="sex",
                    s=100)
plt.title("PCA with Scaled Penguins Data")
plt.savefig("PCA_plot_PC1_vs_PC2_Penguins_scaled_data.png",
                    format='png',dpi=150)
```

```python
sns.plotting_context("talk",font_scale=1.25)
plt.figure(figsize=(8,6))
sns.boxplot(x="species",y="PC1",
            width=0.5,
            data=pc_scaled_df)
plt.xlabel("species", size=14)
plt.title("PC1 vs Species: Penguins Data", size=16)
plt.savefig("PCA_plot_PC1_vs_Species_Penguins_scaled_data.png",
                    format='png',dpi=150)
```

```python
sns.plotting_context("talk",font_scale=1.25)
plt.figure(figsize=(8,6))
sns.boxplot(x="species",y="PC2",
            width=0.5,
            data=pc_scaled_df)
plt.xlabel("species", size=14)
plt.title("PC2 vs Species: Penguins Data", size=16)
plt.savefig("PCA_plot_PC2_vs_Species_Penguins_scaled_data.png",
                    format='png',dpi=150)
```
